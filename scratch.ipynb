```python
from DataTable import DataTable
from Object import Planet, Star

exo = DataTable('exoplanets')
```

```python
import astropy.constants as ac
dt = 1
min_dep = 100 # ppm
for d in exo.table:
    pl = Planet(d)
    transits = pl.get_transits_string(dt)
    transits_jd = pl.get_transits(dt, 0)
    if len(transits) > 0 and pl.transit_depth is not None and pl.transit_depth > min_dep:
        print(pl.name + " " + str(pl.radius) + " " + str(pl.transit_depth))
        for t, tjd in zip(transits, transits_jd):
            print(t + "\n" + str(tjd))
```

```python
dt = 1
min_dep = 100 # ppm
pl_list = []
for d in exo.table:
    pl = Planet(d)
    transits = pl.get_transits_string(dt)
    transits_jd = pl.get_transits(dt, 0)
    if len(transits) > 0 and pl.transit_depth is not None and pl.transit_depth > min_dep:
        pl_list.append((transits[0], transits, pl))
```

```python
import operator
sorted_planets = sorted(pl_list, key=operator.itemgetter(0))
sorted_planets[0][2]
```

```python
for pl in sorted_planets:
    print(pl.name + " " + str(pl.radius) + " " + str(pl.transit_depth))
    for t, tjd in zip(pl.get_transits_string(), pl.get_transits()):
        print(t + "\n" + str(tjd))
```

```python
from fuzzywuzzy import fuzz
ratios = []
for d in exo.table:
    ratios.append(fuzz.ratio('k2 125 b', Planet(d).name))
```
